```python
import pandas as pd
df=pd.read_csv('H:/JY notebook/练习/car_complain.csv')
df
```

```python
df=df.drop('problem',axis=1).join(df.problem.str.get_dummies(','))
df
```

```python
#数据清洗
def f(x):
    x=x.replace('一汽-大众','一汽大众')
    return x
df['brand']=df['brand'].apply(f)
result=df.groupby(['brand'])['id'].agg(['count'])
result
```

```python
result2=df.groupby(['car_model'])['id'].agg(['count'])
result2
```

```python
result3=df.groupby(['brand'])['car_model'].nunique().sort_values(ascending=False)
print(result3)
```

```python
result_final=result.merge(result3,left_index=True,right_index=True,how='left')
data=result_final['count']/result_final['car_model']
data.sort_values(ascending=False)
```
